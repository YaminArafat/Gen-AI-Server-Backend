```python
from diffusers import StableDiffusionPipeline, DiffusionPipeline
import torch
```

```python
high_noise_frac = 0.8
def generate_single_word_image(word: str, output_path: str):
   
    model_id = "stabilityai/stable-diffusion-xl-base-1.0"
    pipe = DiffusionPipeline.from_pretrained(
        model_id,
        torch_dtype=torch.float16, variant="fp16", use_safetensors=True
    ).to("cuda")

    prompt_text = word
    
    image = pipe(
        prompt_text,
        num_inference_steps=30,      
        denoising_end=high_noise_frac,
        output_type="latent",
    ).images

    image.save(output_path)
    return output_path
```

```python
word = "tank"
out_file = f"C:\\Users\\User\\Documents\\Work\\Config Generation Server\\Generated_background_images\\generated_bg_{word}.png"
generate_single_word_image(word, out_file)
print("Saved to", out_file)
```
