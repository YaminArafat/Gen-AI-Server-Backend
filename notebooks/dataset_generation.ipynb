```python


# Map number words to integers
word2num = {
    "zero":0,"none":0,"one":1,"two":2,"three":3,"four":4,
    "five":5,"six":6,"seven":7,"eight":8
}

phrase_map = {
    # Health & Wellness
    "heart rate":             ["edge_hr", "circle_hr"],
    "blood oxygen":           ["edge_blood_oxygen", "circle_blood_oxygen"],
    "sleep coaching":         ["edge_sleep_coaching", "circle_sleep_coaching"],
    "stress level":           ["edge_stress", "circle_stress"],
    "daily steps":            ["edge_steps", "circle_steps"],
    "step count":             ["edge_steps", "circle_steps"],
    "hydration":              ["box_water", "circle_water"],
    "breathing exercises":    ["box_breathe", "circle_breathe"],
    "food intake":            ["box_food", "circle_food"],
    "body composition":       ["edge_body_composition", "circle_body_composition"],
    "vitality score":         ["box_vitality_score", "circle_vitality"],
    "women's health":         ["edge_women_health", "circle_women_health"],
    "blood pressure":         ["edge_blood_pressure", "circle_blood_pressure"],
    "ecg":                    ["edge_ecg", "circle_ecg"],
    "weekly steps":           ["arc_weekly_steps"],
    "workout this week":      ["arc_workout_this_week", "box_workout_this_week"],
    "weekly activity":        ["arc_weekly_activity", "box_weekly_activity"],

    # Fitness & Exercise
    "running":                ["edge_running", "circle_running"],
    "cycling":                ["edge_cycling", "circle_cycling"],
    "hiking":                 ["edge_hiking", "circle_hiking"],
    "swimming":               ["edge_swimming", "circle_swimming"],
    "treadmill":              ["edge_treadmill", "circle_treadmill"],
    "walking":                ["edge_walking", "circle_walking"],
    "circuit training":       ["edge_circuit_training", "circle_circuit_training"],
    "elliptical":             ["edge_elliptical", "circle_elliptical"],
    "bike":                   ["edge_bike", "circle_bike"],
    "track run":              ["edge_track_run", "circle_track_run"],
    "multi sports":           ["edge_multi_sports", "circle_multi_sports"],

    # Weather & Environment
    "chance of rain":         ["edge_chance_of_rain", "circle_chance_of_rain"],
    "chance of rain":         ["arc_chance_of_rain"],
    "detailed weather":       ["arc_detailed_weather", "box_detail_weather"],
    "temperature":            ["arc_temperature", "circle_temperature", "box_temperature"],
    "feel like temperature":  ["edge_feel_like_temp", "circle_feel_like_temp"],
    "fine dust":              ["edge_finedust", "circle_finedust"],
    "ultra fine dust":        ["edge_ultra_finedust", "circle_ultra_finedust"],
    "uv index":               ["edge_uv_index", "circle_uv_index"],
    "sunrise":                ["edge_sunrise", "circle_sunrise"],
    "wind":                   ["edge_wind", "circle_wind"],
    "barometer":              ["edge_barometer", "circle_barometer"],
    "air quality":            ["box_air_quality"],

    # Calendar & Time
    "calendar":               ["box_calendar", "circle_calendar", "circle_calendar_event"],
    "date":                   ["box_date", "arc_date"],
    "moon phase":             ["box_moonphase", "arc_moonphase", "circle_moonphase"],
    "alarm":                  ["box_alarm", "edge_alarm", "circle_alarm"],
    "timer":                  ["box_timer", "edge_timer", "circle_timer"],
    "stopwatch":              ["edge_stopwatch", "circle_stopwatch"],
    "world clock":            ["edge_worldclock", "circle_worldclock"],

    # Utility & Media
    "battery":                ["edge_battery", "circle_battery", "arc_battery"],
    "compass":                ["edge_compass", "circle_compass", "box_compass"],
    "media controls":         ["edge_media_controller", "circle_media_controller"],
    "messages":               ["edge_messages", "circle_messages"],
    "call history":           ["edge_call_history", "circle_call_history"],
    "recent apps":            ["edge_recentApp", "circle_recentApp"],
    "reminder":               ["edge_reminder", "box_reminder", "circle_reminder"],

    # Misc
    "weather":                ["arc_weather", "box_weather", "circle_weather", "edge_weather"],
    "moon":                    ["arc_moonphase", "box_moonphase"],
    "altimeter":              ["circle_altimeter"],
    "steps":                  ["edge_steps", "circle_steps"]
}

# === Explicit Position Mapping ===
# Phrase must include both the item and the position words
explicit_position_map = {
    # edge positions
    "heart rate at top left":       ("edge_hr",          "top_left"),
    "heart rate at top right":      ("edge_hr",          "top_right"),
    "steps at bottom left":         ("edge_steps",       "bottom_left"),
    "steps at bottom right":        ("edge_steps",       "bottom_right"),

    # circle positions
    "temperature at left":          ("circle_temperature","left"),
    "humidity at right":            ("circle_finedust",   "right"),
    "uv index at top":              ("circle_uv_index",   "top"),
    "chance of rain at bottom":     ("circle_chance_of_rain","bottom"),

    # box positions
    "calendar at middle":           ("box_calendar",     "middle"),
    "alarm at bottom":              ("box_alarm",        "bottom"),

    # arc (always top if present)
    "battery arc at top":           ("arc_battery",      "top"),
    "moon phase arc at top":        ("arc_moonphase",    "top")
}

# Helper to extract requested count for edges/circles
def extract_requested_count(text, keyword):
    lower = text.lower()
    # digits
    m = re.search(rf"\b(\d+)\s+{keyword}\b", lower)
    if m:
        return int(m.group(1))
    # words
    for w, n in word2num.items():
        if re.search(rf"\b{w}\s+{keyword}\b", lower):
            return n
    return None

def extract_color_by_context(text, context):
    # Look for patterns like: "context color is blue" or "context color blue"
    pattern = rf"{context} color(?: is)? (\w+)"
    match = re.search(pattern, text)
    if match:
        color = match.group(1)
        if color in background_colors:  # Validate it's a known color
            return color
    return None

# Build config respecting all rules
def make_config_config(text):
    lower = text.lower()
    # Background: theme > mentioned color > random
    bg_color = extract_color_by_context(lower, "background")
    cc_color = extract_color_by_context(lower, "widget")

    theme = next((t for t in image_themes if t in lower), None)
    if theme:
        background = {"color": None, "image": f"/image/path/to/ai_generated_{theme.replace(' ','_')}.png"}
    elif bg_color:
        background = {"color": bg_color, "image": None}
    else:
        background = {"color": random.choice(background_colors), "image": None}

    # widget color
    if cc_color and cc_color != background["color"]:
        widget_color = cc_color
    else:
        widget_color = random.choice([c for c in widget_colors if c != background["color"]])

    # Clock & Dial
    if "analog" in lower: clock_type="analog"
    elif "digital" in lower: clock_type="digital"
    else: clock_type=random.choice(["analog","digital"])
    clock_pos = "center" if clock_type=="analog" else random.choice(clock_positions)
    dial = {"type": random.choice(dial_types)} if clock_type=="analog" else None

    # Determine widget color distinct from background
    # comp_color_choices = [c for c in widget_colors if c != background["color"]]
    # comp_color = comp_color_choices[0] if comp_color_choices else random.choice(widget_colors)

    # Collect requested items in order
    requested = []
    # 1) Explicit count for edges
    edge_req = extract_requested_count(text, "edges")
    if edge_req is None and "all edges" in lower: edge_req=4
    # 2) Explicit positions
    for phrase, (sid, pos) in explicit_position_map.items():
        if phrase in lower:
            style = sid.split("_",1)[0]
            requested.append({"name": sid, "type": style, "position": pos})
    # 3) Phrase-based items
    for phrase, sids in phrase_map.items():
        if phrase in lower:
            for sid in sids:
                style = sid.split("_",1)[0]
                requested.append({"name": sid, "type": style, "position": None})

    # 4) Build initial items list
    items = []
    # Handle edges explicitly
    if edge_req is not None:
        # clamp to allowed {0,2,4}
        cnt = 4 if edge_req>=4 else (2 if edge_req>=2 else 0)
        for pos in positions["edge"][:cnt]:
            sel = random.choice(widgets["edge"])
            items.append({"name": sel, "type":"edge", "position": pos})
    # Add other requested, deduplicating by name+position
    seen = set()
    for req in requested:
        pos = req["position"] or random.choice(positions[req["type"]])
        key = (req["name"], pos)
        if key not in seen:
            seen.add(key)
            items.append({"name": req["name"], "type": req["type"], "position": pos})

    # 5) Clamp cardinalities
    # Group
    by_type = {t: [i for i in items if i["type"]==t] for t in positions}
    # box max1, arc max1
    by_type["box"] = by_type["box"][:1]
    by_type["arc"] = by_type["arc"][:1]
    # edge 0/2/4
    e = len(by_type["edge"])
    if e not in (0,2,4):
        by_type["edge"] = by_type["edge"][:4] if e>2 else []
    # circle per clock type
    c = len(by_type["circle"])
    if clock_type=="digital" and c>3:
        by_type["circle"]=by_type["circle"][:3]
    if clock_type=="analog" and dial and c>0:
        by_type["circle"]=[]
    if clock_type=="analog" and not dial and c>4:
        by_type["circle"]=by_type["circle"][:4]
    # merge back
    items = by_type["edge"] + by_type["circle"] + by_type["box"] + by_type["arc"]

    # 6) Fallback only if no items and not minimal
    if not items and "minimal" not in lower:
        # 2-3 random from circle/edge
        for _ in range(random.randint(2,3)):
            s = random.choice(list(widgets))
            n = random.choice(widgets[s])
            pos = random.choice(positions[s])
            items.append({"name":n,"type":s,"position":pos})

    # Assemble config dict
    wf = {"config": {"Background": background}}
    if dial: wf["config"]["Dial"] = dial
    if items:
        wf["config"]["widget"] = {"color": widget_color, "Item": items}
    wf["config"]["Clock"] = {"type": clock_type, "color": widget_color, "position": clock_pos}
    return wf

# === Dataset Generation ===
OUTPUT_PATH = Path("../datasets/config_dataset_10000.jsonl")
TARGET = 10000
seen_inputs = set()

with OUTPUT_PATH.open("w", encoding="utf-8") as f:
    while len(seen_inputs) < TARGET:
        category = random.choice(list(templates))
        seed = random.choice(templates[category])
        # Paraphrase: append 1-2 sentences
        extra = " ".join(fake.sentences(nb=random.randint(1,2)))
        input_text = f"{seed} {extra}".strip()
        words = input_text.split()
        if len(words) < 10 or len(words) > 256 or input_text in seen_inputs:
            continue
        seen_inputs.add(input_text)
        config = make_config_config(input_text)
        f.write(json.dumps({"input": input_text, "output": config}) + "\n")

print(f"Generated {len(seen_inputs)} samples at {OUTPUT_PATH.resolve()}")

```

```python
# === Helper: Build config Config ===
def make_config_config(text):
    lower = text.lower()
    # Background
    theme = next((t for t in image_themes if t in lower), None)
    background = {"color": None, "image": f"/image/path/to/ai_generated_{theme.replace(' ','_')}.png"} if theme else {"color": random.choice(background_colors), "image": None}
    # Clock
    if "analog" in lower: ct="analog"
    elif "digital" in lower: ct="digital"
    else: ct=random.choice(["analog","digital"])
    cp = "center" if ct=="analog" else random.choice(clock_positions)
    dial = {"type": random.choice(dial_types)} if ct=="analog" else None
    # widget color
    cc = random.choice([c for c in widget_colors if c!=background["color"]])
    # Parse explicit mentions
    items=[]
    for style,names in widgets.items():
        for name in names:
            if name.split("_",1)[1] in lower:
                items.append({"name":name,"type":style,"position":random.choice(positions[style])})
    # Fallback extras
    if "minimal" not in lower and not items:
        # add random 2–3
        for _ in range(random.randint(2,3)):
            s=random.choice(list(widgets))
            n=random.choice(widgets[s])
            items.append({"name":n,"type":s,"position":random.choice(positions[s])})
    # Assemble
    wf={"config":{"Background":background}}
    if dial: wf["config"]["Dial"]=dial
    if items: wf["config"]["widget"]={"color":cc,"Item":items}
    wf["config"]["Clock"]={"type":ct,"color":cc,"position":cp}
    return wf
```

```python
# === Generation Loop ===
OUTPUT_PATH = Path("../datasets/config_randomized_dataset_10000.jsonl")
TARGET = 10000
seen = set()

with OUTPUT_PATH.open("w", encoding="utf-8") as out:
    while len(seen) < TARGET:
        cat = random.choice(list(templates))
        seed = random.choice(templates[cat])
        # Paraphrase seed with 1–3 Faker sentences
        extra = " ".join(fake.paragraph(nb_sentences=random.randint(1,3)).split("\n"))
        inp = f"{seed} {extra}"
        words = inp.split()
        if len(words)<10 or len(words)>256 or inp in seen:
            continue
        seen.add(inp)
        cfg = make_config_config(inp)
        record = {"input": inp, "output": cfg}
        out.write(json.dumps(record) + "\n")

print(f" Wrote {len(seen)} samples to {OUTPUT_PATH.resolve()}")
```
